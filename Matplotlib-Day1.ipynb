```python
# req :  Sales data

# Revenue Vs Sales ====> 
# x axis ===> Sales
# y axis ===> Revenue
# title of the graph ====> Performance calculator of sales
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]

plt.title('Performance calculator of Sales')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['this is x and y', 'this is x and z'])
plt.show()
```

```python
Compare population between US and China :
```

```python
plt.plot(us.year,us.population)
plt.plot(china.year,china.population)
plt.xlabel('population')
# Set the y axis label of the current axis.
plt.ylabel('year')
# Set a title of the current axes.
plt.title('US Vs China - population ')
# show a legend on the plot
plt.legend(['United States','China'])
plt.show()
```
