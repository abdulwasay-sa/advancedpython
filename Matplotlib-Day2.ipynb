```python
Pie chart making using matplotlib:
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Basic approach : 
```

```python
y =np.array([35,25,25,15])
```

```python
plt.pie(y)
plt.show()
```

```python
# req.: to add labels to the pie chart
```

```python
mylabels = ['Apples','Bananas','Cherries','Dates']
```

```python
plt.pie(y,labels = mylabels)
plt.show()
```

```python
# how to control the angles of the pie chart....default it will start at 0.
```

```python
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
# explode the pie chart
```

```python
myexplode = [0.2,0.3,0,0]
```

```python
plt.pie(y, explode = myexplode, labels = mylabels)
plt.show()
```

```python
# shadow the exploded pie's and add legend to the pie chart
```

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```

```python
Creating the bars with matplotlib:
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
# custom colour :
```

```python
plt.bar(x,y, color = 'red')
plt.show()
```

```python
plt.bar(x,y, color = 'hotpink')
plt.show()
```

```python
# Width control : 
```

```python
plt.bar(x,y, width= 0.2)
plt.show()
```
