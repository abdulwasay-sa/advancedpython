```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv('countries.csv')
df
```

```python
# Compare Five countries & plot their population :  Bahrain, Kuwait, Oman, Saudi Arabia, Yemen, Rep. (gulf countries)
```

```python
bahrain = df[df.country == 'Bahrain']  # filter operation ===> only with Bahrain data
bahrain
```

```python
kuwait = df[df.country == 'Kuwait']  # filter operation ===> only with Kuwait data
kuwait
```

```python
oman = df[df.country == 'Oman']  # filter operation ===> only with Oman data
oman
```

```python
ksa = df[df.country == 'Saudi Arabia']  # filter operation ===> only with Saudi Arabia data
ksa
```

```python
yemen = df[df.country == 'Yemen, Rep.']  # filter operation ===> only with Yemen, Rep. data
yemen
```

```python
plt.plot(bahrain.year,bahrain.population)
plt.plot(kuwait.year,kuwait.population)
plt.plot(oman.year,oman.population)
plt.plot(ksa.year,ksa.population)
plt.plot(yemen.year,yemen.population)

plt.xlabel('population')
# Set the y axis label of the current axis.
plt.ylabel('year')
# Set a title of the current axes.
plt.title('Gulf Countries - population comparision')
# show a legend on the plot
plt.legend(['Bahrain', 'Kuwait', 'Oman', 'Saudi Arabia','Yemen'])
# Display a figure.
plt.show()
```
